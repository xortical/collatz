# Function to Find how many number have collatz trees whose hailstone numbers are less than MAX

```python
def hailstonelessthanMAX(MAX):
    n_list = range(1,MAX+1)
    count = 0


    for n in n_list:
        num = n
        hailstones = []
        lessThanMaxFlag = True
        while not n == 1:
            if n % 2 == 0:
                n /= 2
            else:
                n = 3*n + 1
            hailstones.append(n)
            if n>MAX:
                lessThanMaxFlag = False
                break
        if lessThanMaxFlag:
            count+=1
    return(count/MAX)

```

```python
percent = []
for i in range(1,100000):
    percent.append(hailstonelessthanMAX(i))
```

```python
plt.plot(percent)
plt.grid()
```
